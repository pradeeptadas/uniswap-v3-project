# Covalent Data

```python
import requests
import pandas as pd
```

```python
API_KEY = 'ckey_5fb2f250010f42438610a820d45'
```

```python
def unravel_token(df, token):
    df1 = df[token].apply(pd.Series)
    df1.columns = [f'{token}__' + i for i in df1.columns]
    df2 = pd.concat([df1, df.drop([token], axis=1)], axis=1)
    return df2
```

## pool data

```python
def get_pools_url(chain_id=1, page_size=1000):
    api_url = 'https://api.covalenthq.com'
    endpoint = f'/v1/{chain_id}/uniswap_v3/pools/?page-size={page_size}'
    url = api_url + endpoint
    return url
```

```python
response = requests.get(get_pools_url(page_size=5000), auth=(API_KEY,''))
data = response.json()
```

```python
if data['error']:
    raise Exception(data['error_message'])
```

```python
pool_df = pd.DataFrame(data['data']['items'])
```

```python
pool_df = unravel_token(pool_df, 'token_1')
pool_df = unravel_token(pool_df, 'token_0')
```
